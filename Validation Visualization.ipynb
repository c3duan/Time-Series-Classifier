```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load(filename):
    data_dict = {}
    with open(filename, 'r') as f:
        for i, line in enumerate(f.readlines()):
            if i == 0:
             continue
            line = line.rstrip().split(',')
            value, rank, group = float(line[0]), int(line[1]), int(line[2])
            if group not in data_dict:
                data_dict[group] = []
                for _ in range(6):
                    data_dict[group].append([])

            data_dict[group][rank].append(value)
    return data_dict        
```

```python
siamese_data_dict = load('SIAMESE_validation_data.csv')
cosim_data_dict = load('COSINE_SIM_validation_data.csv')
DTW_data_dict = load('DTW_validation_data.csv')
```

```python
def visualize(data_dict, offset=0, rank=1):
    fig, axes = plt.subplots(nrows=10, ncols=2, figsize=(16, 16))
    for i in range(10):
        group = i + offset
        x = np.arange(1, len(data_dict[group][0]) + 1)
        axes[i, 0].plot(x, data_dict[group][0])
        axes[i, 0].set_title('Original')
        axes[i, 1].plot(x, data_dict[group][rank])
        axes[i, 1].set_title('Rank 1')

    fig.tight_layout()
    plt.show()
```

# Siamese Network

```python
for i in range(0, 40, 10):
    visualize(siamese_data_dict, i)
```

# Cosine Similarity

```python
for i in range(0, 40, 10):
    visualize(cosim_data_dict, i)
```

# DTW

```python
for i in range(0, 40, 10):
    visualize(DTW_data_dict, i, 2)
```
